```python
import pandas as pd
```

```python
df = pd.read_excel('평택매매2022.xlsx')
df = df[15:]
col = list(df.iloc[0])
df.columns = col
df = df[1:]
```

```python
for i in [2020,2021]:
    df_plus = pd.read_excel(f'평택매매{i}.xlsx',thousands=',')
    df_plus = df_plus[15:]
    col = list(df_plus.iloc[0])
    df_plus.columns = col
    df_plus = df_plus[1:]
    df = df.append(df_plus)
    
df
```

```python
df.columns
```

```python
df_pick = df[['단지명','전용면적(㎡)', '계약년월','거래금액(만원)','층', '건축년도','도로명']]
df_pick = df_pick.sort_values(by=['단지명','계약년월'])
df_pick.to_csv('평택매매.csv', encoding = 'utf-8-sig')
df_pick = df_pick.groupby(['단지명','전용면적(㎡)','계약년월']).max()
df_pick
```

### 전월세 데이터 정리

```python
df_rent = pd.read_excel('평택전세2022.xlsx')
df_rent = df_rent[15:]
col = list(df_rent.iloc[0])
df_rent.columns = col
df_rent = df_rent[1:]
```

```python
for i in [2020,2021]:
    df_plus = pd.read_excel(f'평택전세{i}.xlsx',thousands=',')
    df_plus = df_plus[15:]
    col = list(df_plus.iloc[0])
    df_plus.columns = col
    df_plus = df_plus[1:]
    df_rent = df_rent.append(df_plus)
    
df_rent.shape
```

```python
df_rent.columns
```

```python
month = df_rent[df_rent['전월세구분'] == '월세'].index
df_rent = df_rent.drop(month)
df_rent
```

```python
df_rent_pick = df_rent[['단지명','전용면적(㎡)', '계약년월','보증금(만원)','층', '건축년도','도로명']]
df_rent_pick = df_rent_pick.sort_values(by=['단지명','계약년월'])
df_rent_pick = df_rent_pick.groupby(['단지명','전용면적(㎡)','계약년월']).max()
df_rent_pick
```

```python
df = pd.merge(df_pick,df_rent_pick,on=['단지명','전용면적(㎡)','계약년월'])
df.pop('건축년도_x')
df.pop('도로명_x')
df = df.sort_values(by=['단지명','계약년월'])
df.head(5)
```

```python
df[['보증금(만원)', '거래금액(만원)']] = df[['보증금(만원)', '거래금액(만원)']].apply(pd.to_numeric)
df['전세가율'] = df['보증금(만원)'] / df['거래금액(만원)']
df['전세가율']
```
