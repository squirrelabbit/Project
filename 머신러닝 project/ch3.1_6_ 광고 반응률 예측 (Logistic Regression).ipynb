# ch3.4 Missing Value 확인 및 처리

```python
import pandas as pd
import numpy as np

# 데이터 시각화 모듈
import matplotlib.pyplot as plt
import seaborn as sns

# 워닝 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('advertising.csv')
```

## 결측치 확인

```python
data.info()
```

```python
# missing value(결측치) 확인
data.isna()
```

```python
# missing value(결측치) 갯수 확인
data.isna().sum()
```

```python
# missing value(결측치) 비율 확인
data.isna().sum() / len(data)
```

```python
len(data)
```

## 결측치 처리하기

```python
# 결측치 제거
data.dropna()
```

```python
# 컬럼 삭제
data.drop('Age', axis=1)
```

```python
# mean : 평균값
round(data['Age'].mean())
```

```python
# median : 중간값 - 이상치의 영향을 없앨 수 있음.
data['Age'].median()
```

```python
# 결측치를 평균값으로 채워넣기
data = data.fillna(round(data['Age'].mean()))
```

```python
data
```

```python
# 결측값이 더 이상 없음 확인
data.isna().sum()
```

# ch3.5 Train / Test 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[['Daily Time Spent on Site','Age','Area Income', 'Daily Internet Usage','Male']]
y = data['Clicked on Ad']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
X_train
```

```python
X_test
```

# ch3.6 로지스틱 리그레션 모델 만들기

```python
# 사이킷런의 로지스틱 리그레션 가져오기
from sklearn.linear_model import LogisticRegression
```

```python
# LogisticRegression 모델 변수 생성
model = LogisticRegression()
```

## 모델 학습

```python
# 모델 학습
model.fit(X_train, y_train)
```

```python
# Coefficient 확인
model.coef_
```

## test 데이터 예측

```python
# predict : 예측
pred = model.predict(X_test)
```

```python
print(len(pred))
pred
```

```python
print(len(y_test))
y_test
```

## 모델 평가

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```
