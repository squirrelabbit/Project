```python
data_ages = pd.read_excel('Report.xls')
data_ages.info()
```

```python
data_ages['청년(20~34)']=data_ages.iloc[:,4:8].sum(axis=1)
data_ages['중장년(35~64)']=data_ages.iloc[:,8:14].sum(axis=1)
data_ages['노년(65~)']=data_ages.iloc[:,14:19].sum(axis=1)
data_ages
```

```python
data_ages_cut = data_ages.iloc[:,[1,2,3,19,20,21]] 
data_ages_cut
```

```python
data_ages_gu =data_ages_cut.loc[data_ages_cut.성별 == "계"]
data_ages_gu=data_ages_gu[1:]
data_ages_gu
```

```python
data_ages_gu["청년비중"] = data_ages_gu["청년(20~34)"]/data_ages_gu["합계"]
data_ages_gu["중장년비중"] = data_ages_gu["중장년(35~64)"]/data_ages_gu["합계"]
data_ages_gu["노년비중"] = data_ages_gu["노년(65~)"]/data_ages_gu["합계"]
data_ages_gu
```

```python
df_young = data_ages_gu.sort_values(by = "청년(20~34)",ascending = False)
df_young
```
